## filter the converted customer

```python
df_true = df[df['convert_TF'] == True]
df_true_t1 = df_true[df_true['tier'] == 1]
df_true_t2 = df_true[df_true['tier'] == 2]
df_true_t3 = df_true[df_true['tier'] == 3]
```

## convert dataframe to dict

```python
# to dictionary
df_true_t1_dic = df_true_t1.to_dict(orient='records')
df_true_t2_dic = df_true_t2.to_dict(orient='records')
df_true_t3_dic = df_true_t3.to_dict(orient='records')
```

```python
import math

# dic for tier 1
df_true_t1_dic
df_true_t1_dic_channel = []

for item in df_true_t1_dic:
    list_touch = ['touch1', 'touch2', 'touch3', 'touch4','touch5']
    list1 = []
    for t in list_touch:
        if type(item[t]) == float:
            pass
        else:
            list1.append(item[t])
        
    df_true_t1_dic_channel.append(list1)

# dic for tier 2
df_true_t2_dic
df_true_t2_dic_channel = []
for item in df_true_t2_dic:
    list_touch = ['touch1', 'touch2', 'touch3', 'touch4','touch5']
    list1 = []
    for t in list_touch:
        if type(item[t]) == float:
            pass
        else:
            list1.append(item[t])
        
    df_true_t2_dic_channel.append(list1) 

# dic for tier 3
df_true_t3_dic
df_true_t3_dic_channel = []
for item in df_true_t3_dic:
    list_touch = ['touch1', 'touch2', 'touch3', 'touch4','touch5']
    list1 = []
    for t in list_touch:
        if type(item[t]) == float:
            pass
        else:
            list1.append(item[t])
        
    df_true_t3_dic_channel.append(list1) 
```

## credit distributed to each channel

```python
t1_channel_count = {'social':0, 'organic_search':0, 'referral':0, 'email':0, \
                           'paid_search':0, 'display':0, 'direct':0}
t2_channel_count = {'social':0, 'organic_search':0, 'referral':0, 'email':0, \
                           'paid_search':0, 'display':0, 'direct':0}
t3_channel_count = {'social':0, 'organic_search':0, 'referral':0, 'email':0, \
                           'paid_search':0, 'display':0, 'direct':0}

for item in df_true_t1_dic_channel:
    
    # if there is only one channel for this customer journey
    # will attribute all credit to this channel
    if len(item) == 1:
        t1_channel_count[item[0]] += 1
    
    
    # if there is two channels for this customer journey
    # will attribute half credit to both of the channel
    elif len(item) == 2:
        for i in item:
            t1_channel_count[i] += 0.5
        
    # if there is more than two channel,
    # will practice position based method
    else:
        m_credit = 0.2/(len(item) - 2)
        t1_channel_count[item[0]] += 0.4
        t1_channel_count[item[-1]] += 0.4
        
        for i, v in enumerate(item):

            if i != 0 and i != (len(item)-1):
                t1_channel_count[item[i]] += m_credit
            else:
                pass
            
```

```python
for item in df_true_t2_dic_channel:
    
    # if there is only one channel for this customer journey
    # will attribute all credit to this channel
    if len(item) == 1:
        t2_channel_count[item[0]] += 1
    
    
    # if there is two channels for this customer journey
    # will attribute half credit to both of the channel
    elif len(item) == 2:
        for i in item:
            t2_channel_count[i] += 0.5
        
    # if there is more than two channel,
    # will practice position based method
    else:
        m_credit = 0.2/(len(item) - 2)
        t2_channel_count[item[0]] += 0.4
        t2_channel_count[item[-1]] += 0.4
        
        for i, v in enumerate(item):

            if i != 0 and i != (len(item)-1):
                t2_channel_count[item[i]] += m_credit
            else:
                pass
```

```python
for item in df_true_t3_dic_channel:
    
    # if there is only one channel for this customer journey
    # will attribute all credit to this channel
    if len(item) == 1:
        t3_channel_count[item[0]] += 1
    
    
    # if there is two channels for this customer journey
    # will attribute half credit to both of the channel
    elif len(item) == 2:
        for i in item:
            t3_channel_count[i] += 0.5
        
    # if there is more than two channel,
    # will practice position based method
    else:
        m_credit = 0.2/(len(item) - 2)
        t3_channel_count[item[0]] += 0.4
        t3_channel_count[item[-1]] += 0.4
        
        for i, v in enumerate(item):

            if i != 0 and i != (len(item)-1):
                t3_channel_count[item[i]] += m_credit
            else:
                pass
```
