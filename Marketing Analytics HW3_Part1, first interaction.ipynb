```python
import pandas as pd
```

```python
df = pd.read_csv('attribution_allocation_student_data (1).csv')
cost = pd.read_csv('channel_spend_student_data.csv')
```

```python
df1 = df[df['tier']==1]
df1 = df1[df1['convert_TF']==True]
d1 = pd.DataFrame(df1['touch1'].value_counts())
d1.rename(columns={'touch1':'tier1'},inplace = True)
d1
```

```python
df2 = df[df['tier']==2]
df2 = df2[df2['convert_TF']==True]
df2['touch1'].value_counts()
d2 = pd.DataFrame(df2['touch1'].value_counts())
d2.rename(columns={'touch1':'tier2'},inplace = True)
d2.reset_index()
d2
```

```python
df3 = df[df['tier']==3]
df3 = df3[df3['convert_TF']==True]
df3['touch1'].value_counts()
d3 = pd.DataFrame(df3['touch1'].value_counts())
d3.rename(columns={'touch1':'tier3'},inplace = True)
d3
```

```python
cost.set_index(['Unnamed: 0'],inplace=True)
```

```python
cost
```

```python
tier1 = eval(cost.loc['tier1','0'])
tier2 = eval(cost.loc['tier2','0'])
tier3 = eval(cost.loc['tier3','0'])
```

```python
tier1
```

```python
for k in list(tier1.keys()):
    #d.loc[k, 'ptier1'] = tier1[k]
    d.loc[k,'AvgCACtier1'] = tier1[k]/d.loc[k,'tier1']
for k in list(tier2.keys()): 
    d.loc[k,'AvgCACtier2'] = tier2[k]/d.loc[k,'tier2']
for k in list(tier3.keys()):
    d.loc[k,'AvgCACtier3'] = tier3[k]/d.loc[k,'tier3']
```

```python
for k in list(tier1.keys()):
    #d.loc[k, 'ptier1'] = tier1[k]
    d.loc[k,'Ctier1'] = tier1[k]
for k in list(tier2.keys()): 
    d.loc[k,'Ctier2'] = tier2[k]
for k in list(tier3.keys()):
    d.loc[k,'Ctier3'] = tier3[k]
d['totalC'] = d['Ctier1'] + d['Ctier2'] + d['Ctier3']
```

```python
d['totalCount'] = d['tier1']+d['tier2']+d['tier3']
d
```

```python
d['overallCost'] = d['totalC']/d['totalCount']
d
```
